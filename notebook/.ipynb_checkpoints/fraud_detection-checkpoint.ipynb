```python
train_data = pd.read_csv('../data/fraud_train.csv')
train_data
```

```python
# do some preprocessing !
def remove_outlier(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return series[(series < (Q1-1.5*IQR)) | (series > (Q3+1.5*IQR))].index.to_list()
idx = []
for  i in train_data.columns[:30]:
    idx.extend(remove_outlier(train_data[train_data.Class==0][i]))
train_data.drop(index=idx,inplace=True)
train_data=train_data.reset_index().drop('index',axis=1)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV ,train_test_split
from scipy import stats

X= train_data.drop('Class',axis=1)
y= train_data.Class

param_grid = {'C': stats.uniform(2, 10), 
              'gamma': stats.uniform(0.1, 1),
              'kernel': ['rbf', 'linear']} 
clf = RandomizedSearchCV(SVC(), param_grid, verbose=3)
clf.fit(X , y)
params = clf.best_params_
```

```python
model = SVC(C=params['C'] , gamma= params['gamma'] , kernel = params['kernel'])
X_train, X_test, y_train,y_test= train_test_split(X,y ,test_size=0.2 , shuffle=True , random_state=43)
model.fit(X_train , y_train)
y_predict = model.predict(X_test)
```

```python
# evaluate your model
from sklearn.metrics import f1_score, classification_report
score= f1_score(y_test,y_predict)
score
```
